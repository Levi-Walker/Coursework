# NLP with Machine Learning

# Ham or Spam

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Dataset source: https://www.kaggle.com/datasets/venky73/spam-mails-dataset
df = pd.read_csv("datasets/SpamMailsDataset/spam_ham_dataset.csv")
```

```python
df.head()
```

```python
emails = df[["text","label_num"]]
```

```python
emails.head()
```

```python
emails.shape
```

```python
emails['label_num'].value_counts()
```

```python
texts = emails['text']
targets = emails['label_num']
```

```python
targets.head()
```

## Vectorizer

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)
```

```python
print(X.shape)
```

```python
print(X[1])
```

## Dataset split

```python
X_train, X_test, y_train, y_test = train_test_split(X, targets, test_size = 0.3, random_state=3)
```

```python
print(f"Train shape: {X_train.shape}")
print(f"Test shape: {X_test.shape}")
```

## Train a ML model (Random Forest)

```python
forest = RandomForestClassifier(n_estimators=100)
forest.fit(X_train,y_train)
```

## Predictions

```python
predict = forest.predict(X_test)
```

```python
print(predict[:10])
print(y_test[:10])
```

```python
confusion_matrix(y_test, predict)
```

```python
print(f"Accuracy: {accuracy_score(y_test, predict)}")
```

```python
print(metrics.classification_report(y_test, predict))
```

## Saving and loading the model

```python
import joblib
```

```python
joblib.dump(forest, 'spamHam.joblib')
joblib.dump(vectorizer, 'vectorizer_spam.joblib')
print("Model and vectorizer saved")
```

```python
model = joblib.load('spamHam.joblib')
loaded_vectorizer = joblib.load('vectorizer_spam.joblib')
print("Model and vectorizer loaded")
```

## Using the trained model in real samples

```python
file = "datasets/SpamMailsDataset/spam2.txt"
```

```python
with open(file,'r') as text:
    email = text.read()

print(email)
```

```python
email_transformed = loaded_vectorizer.transform([email])
```

```python
predict = model.predict(email_transformed)
predict
```

```python
if predict[0] == 1:
    print("SPAM")
else:
    print("HAM")
```
