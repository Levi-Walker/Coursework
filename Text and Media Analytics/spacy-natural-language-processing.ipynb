### lex_id vs orth
In spaCy, token.lex_id (hash) is a normalized identifier based on the lemma of the word, making it case-insensitive and unaffected by variations in text. In contrast, token.orth is a hash of the exact form of the word as it appears in the text, preserving case and formatting differences.

```python
words = nlp(text = "Run run RUN")

for token in words:
    print(f"Word: {token.text} - Hash: {token.lex_id}, orth: {token.orth}")
```

## **Part-of-speech tagging**
Part-of-speech (POS) tagging assigns a grammatical category, such as noun, verb, or adjective, to each token. This helps in understanding the structure and meaning of sentences.

</div>

**Text**: The original word text.<br>
**Lemma**: The base form of the word.<br>
**POS**: The simple UPOS part-of-speech tag.<br>
**Tag**: The detailed part-of-speech tag.<br>
**Dep**: Syntactic dependency, i.e. the relation between tokens.<br>
**Shape**: The word shape – capitalization, punctuation, digits.<br>
**is alpha**: Is the token an alpha character?<br>
**is stop**: Is the token part of a stop list, i.e. the most common words of the language?<br>

## **Similarity**
spaCy allows you to measure the similarity between texts, tokens, or documents using vector-based methods. It leverages pre-trained word embeddings, calculating cosine similarity to compare the semantic proximity of words or phrases. For example, doc1.similarity(doc2) returns a similarity score between 0 and 1, where higher scores indicate greater similarity.

</div>

### Improving Similarity Performance

We can enhance performance and accuracy by preprocessing your text. One effective strategy is to focus on shorter sentences or phrases instead of large chunks of text. This reduces noise and ensures that the similarity computation captures the most relevant information.  

Additionally, removing **stopwords** (common words like "the," "is," or "and") can significantly improve results. 
.

## **Match**
The Matcher in spaCy is a powerful tool for identifying patterns in text based on token attributes like text, POS tags, or dependency relations. Using rule-based matching, you can define patterns (e.g., sequences of words) and find instances of these patterns in a document. It’s ideal for tasks like extracting specific phrases or entities not covered by default NLP pipelines.

</div>

## SpaCy Matcher Operators

| Operator | Description                              | Example          |
|----------|------------------------------------------|------------------|
| `!`      | Negation: the token must not match.      | `"OP": "!"`      |
| `?`      | Optional: the token can appear 0 or 1 time. | `"OP": "?"`      |
| `+`      | One or more: the token must appear at least once. | `"OP": "+"` |
| `*`      | Zero or more: the token can appear any number of times, including none. | `"OP": "*"` |


### Token Attributes for spaCy Matcher

| Attribute    | Description                                                        |
|--------------|--------------------------------------------------------------------|
| `ORTH`       | The exact verbatim text of the token.                              |
| `TEXT`       | The literal text of the token (same as `ORTH`).                    |
| `LOWER`      | The lowercase form of the token text.                              |
| `LEMMA`      | The lemma (base form) of the token.                                |
| `POS`        | The part-of-speech tag of the token.                               |
| `TAG`        | The fine-grained part-of-speech tag of the token.                  |
| `DEP`        | The syntactic dependency label of the token.                       |
| `SHAPE`      | The shape of the token (e.g., "Xx" for "Apple", "dd" for "42").    |
| `IS_ALPHA`   | Boolean indicating if the token consists of alphabetic characters. |
| `IS_DIGIT`   | Boolean indicating if the token consists of digits.                |
| `IS_TITLE`   | Boolean indicating if the token is title-cased.                    |
| `IS_PUNCT`   | Boolean indicating if the token is punctuation.                    |
| `IS_SPACE`   | Boolean indicating if the token consists of whitespace.            |
| `ENT_TYPE`   | The entity type of the token (e.g., "PERSON", "ORG").              |
| `LENGTH`     | The length of the token in characters.                             |


```python
from typing import Match
from spacy.matcher import Matcher
```

```python
docMatch = nlp("If you need assistance, please call us at (269) 471-7771 or toll-free at (800) 253 2874.")
```

```python
matcher = Matcher(nlp.vocab)
```

```python
phone_pattern = [
    {"ORTH":"("},
    {"SHAPE":"ddd"},
    {"ORTH":")"},
    {"SHAPE":"ddd"},
    {"ORTH":"-", "OP": "?"},
    {"SHAPE":"dddd"},
]
```

```python
matcher.add("phone", [phone_pattern])
matches = matcher(docMatch)
```

```python
for id, start, end in matches:
    print(docMatch[start:end])
```

```python
doc = nlp("The athlete runs every morning, but yesterday she ran and was running late.")
```

```python
pattern = [{"LEMMA":"run"}]
matcher.add("RUN_PATTERN", [pattern])
```

```python
matches = matcher(doc)
for id, start, end in matches:
    print(doc[start:end])
```

```python
doc = nlp(text)

matcher = Matcher(nlp.vocab)

pattern = [
    [{"POS":"PROPN"}, {"POS":"PROPN"}],
    [{"LOWER":"united"}, {"LOWER":"nations"}],
    [{"POS":"VERB"}, {"POS":"NOUN"}]
]

matcher.add("ProperNunPattern", [pattern[0]])
matcher.add("UnitedNationsPattern", [pattern[1]])
matcher.add("VerbNounPattern", [pattern[2]])

matches = matcher(doc)
```

# **Working with the Pipeline**

The SpaCy pipeline is a sequence of components used to process and analyze text efficiently. Each component performs a specific task, such as tokenization, named entity recognition, or dependency parsing. The modular design allows users to customize the pipeline by adding, removing, or modifying components based on their use case.

### Components of the SpaCy Pipeline

| **Component**        | **Description**                                                                 |
|-----------------------|---------------------------------------------------------------------------------|
| **Tokenizer**         | Splits the text into tokens, handling punctuation, whitespace, and special rules.|
| **Tagger**            | Assigns part-of-speech (POS) tags to each token.                                |
| **Parser**            | Analyzes syntactic dependencies and assigns sentence boundaries.                |
| **NER** (Entity Recognizer) | Identifies named entities like people, organizations, dates, etc.                    |
| **TextCategorizer**   | Classifies the text into predefined categories.                                |
| **Lemmatizer**        | Converts words to their base forms (lemmas).                                   |
| **Attribute Ruler**   | Adds or modifies token attributes based on patterns or rules.                  |
| **Sentencizer**       | Splits text into sentences without dependency parsing.                         |
| **Tok2Vec**           | Learns context-sensitive token vectors, often used as input to downstream tasks.|
| **Transformer**       | Utilizes transformer-based embeddings (e.g., BERT) for advanced representations.|
| **Morphologizer**     | Assigns fine-grained morphological features to tokens.                         |
| **EntityLinker**      | Links named entities to a knowledge base (e.g., Wikidata).                     |
| **Custom Components** | User-defined components for specific tasks, such as custom text transformations.|

## **2. Sentiment Analysis (Pedagogical Approach)**
### (Collaborative Work with the Instructor)

Create a function that classifies sentences as positive, negative, or neutral based on the presence of specific words. Analyze the sentiment of a given paragraph and calculate the percentage of each sentiment category.

This exercise uses a basic approach to sentiment analysis by manually defining lists of positive and negative words. It is important to note that this method is **pedagogical** and not commonly used in real-world applications. 

In real scenarios, sentiment analysis typically relies on pre-trained machine learning models or specialized libraries that consider the context and grammar of sentences. This simplified approach helps you understand the basic concept of sentiment classification but lacks the sophistication and accuracy of modern NLP tools.

</div>


```python
def analyze_sentiment(text):
    positive_words = ['amazing', 'great', 'good', 'positive', 'happy', 'excellent']
    negative_words = ['bad', 'terrible', 'sad', 'negative', 'poor', 'horrible']

    # Tokenize the text
    doc = nlp(text)
    sentiment_counts = {'positive': 0, 'negative': 0, 'neutral': 0}


    for token in doc:
        if token.text.lower() in positive_words:
            sentiment_counts['positive'] += 1
        elif token.text.lower() in negative_words:
            sentiment_counts['negative'] += 1
        else:
            sentiment_counts['neutral'] += 1

    total = sum(sentiment_counts.values())
    for sentiment, count in sentiment_counts.items():
        percentage = (count / total) * 100
        print(f"{sentiment.capitalize()}: {percentage:.2f}%")

    
```

```python

example_text = '''
SpaCy is an amazing library. However, learning it can be a bit challenging at times. 
Overall, it is a great tool for NLP.
'''

analyze_sentiment(example_text)
```
