# Loading and understanding images

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = 'images/whale.jpg'
image = cv2.imread(path)
```

```python
plt.imshow(image)
plt.axis("off")
plt.show()
print(image.shape)
```

```python
rgbImage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgbImage)
plt.axis("off")
plt.show()
print(image.shape)
```

```python
grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grayImage, cmap="gray")
plt.axis("off")
plt.show()
print(image.shape)
```

```python
r = rgbImage[:,:,0]
g = rgbImage[:,:,1]
b = rgbImage[:,:,2]

plt.imshow(r, cmap="Reds")
plt.axis("off")
plt.show()

plt.imshow(g, cmap="Greens")
plt.axis("off")
plt.show()

plt.imshow(b, cmap="Blues")
plt.axis("off")
plt.show()
```

## Transformations

```python
crop = rgbImage[700:1200,400:1000]

plt.imshow(crop)
plt.axis("off")
plt.show()
```

## Resize

#### Resizing using a cv2 function

```python
height = image.shape[0]
width = image.shape[1]

proportion = float(height/width)
print(proportion)

newHeight = 400
newWidth = int(newHeight * proportion)
```

```python
resized = cv2.resize(rgbImage, (newHeight, newWidth))

plt.imshow(resized)
plt.show()
```
