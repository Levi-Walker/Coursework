# Feature Scaling Techniques

A comparison of standard, min-max, and robust scaling on synthetic and real-valued features.

```python
# Feature scaling techniques

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler

```

```python
from numpy.random import default_rng

rng = default_rng(0)

rng.random(10)
```

```python
rng.random(10)*6-3
```

```python
df = pd.DataFrame({
    "Col_1" : rng.random(10)*6-3,
    "Col_2" : rng.random(10)*2500-500,
    "Col_3" : rng.random(10)*2,
})
```

```python
df.head()
```

```python
# Standard scaling: (x_i - mean) / standard deviation
```

```python
columns = df.columns
features = df

scaler = StandardScaler().fit(features)
features = scaler.transform(features)


print(features.max())
print(features.min())

scaled_features = pd.DataFrame(features, columns = columns)

scaled_features.head()


```

```python
columns = df.columns
features = df

scaler = MinMaxScaler().fit(features)
features = scaler.transform(features)


print(features.max())
print(features.min())

scaled_features = pd.DataFrame(features, columns = columns)

scaled_features.head()


```

```python
columns = df.columns
features = df

scaler = RobustScaler().fit(features)
features = scaler.transform(features)


print(features.max())
print(features.min())

scaled_features = pd.DataFrame(features, columns = columns)

scaled_features.head()


```

```python
# Haberman dataset
cols = ["Age", "Year", "Nodes", "Class"]

dataset = pd.read_csv('Desktop/haberman+s+survival/haberman.data', names = cols, header=None, sep=",")

dataset
```

```python
# Robust scaling
columns = dataset.drop(columns=["Class"]).columns
features = dataset.drop(columns=["Class"])

scaler = RobustScaler().fit(features)
features = scaler.transform(features)


print(features.max())
print(features.min())

scaled_features = pd.DataFrame(features, columns = columns)

scaled_features.head()
```

```python
# Min-max scaling
columns = dataset.drop(columns=["Class"]).columns
features = dataset.drop(columns=["Class"])

scaler = MinMaxScaler().fit(features)
features = scaler.transform(features)


print(features.max())
print(features.min())

scaled_features = pd.DataFrame(features, columns = columns)

scaled_features.head()
```

```python
# Standard scaling
columns = dataset.drop(columns=["Class"]).columns
features = dataset.drop(columns=["Class"])

scaler = StandardScaler().fit(features)
features = scaler.transform(features)


print(features.max())
print(features.min())

scaled_features = pd.DataFrame(features, columns = columns)

scaled_features.head()
```
