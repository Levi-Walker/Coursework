# Binary Iris Classification with a Neural Network

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

## Dataset

```python
path = ""
dataSet = pd.read_csv(path+'iris.csv')
dataSet = dataSet[:100] # Keep the Iris-setosa and Iris-versicolor classes
# Shuffle the DataFrame
data = dataSet.sample(frac=1).reset_index(drop=True)

data[20:40]
```

```python
data["Species"] = data["Species"].apply(lambda x: x.replace("Iris-setosa", "0"))
data["Species"] = data["Species"].apply(lambda x: x.replace("Iris-versicolor", "1"))
data = data.astype(float)
data[:20]
```

```python
X = data.iloc[:,1:5].values # Feature columns
y = data.iloc[:,5:6].values # Target column
y = y.astype(int)
print(X[:5])
print(y[:5])

print(np.shape(X))
print(np.shape(y))
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    X, 
    y,
    test_size = 0.2,
    random_state=10
)
print(len(data))
print(len(x_train))
print(len(x_test))
```

## TensorFlow model

### Architecture

```python
model = Sequential(
    [ 
        Dense(4, activation = 'relu', name = 'layer1'),
        Dense(1, activation = 'sigmoid', name = 'layer2')    
    ]
)
```

```python
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(0.01),
)

model.fit(
    x_train, y_train, 
    epochs=100
)
```

The `model.summary()` provides a description of the network:

```python
model.summary()
```

```python
loss = model.evaluate(x_test, y_test)
print(loss)
```

The parameter counts shown in the summary correspond to the number of elements in the weight and bias arrays as shown below.

```python
L1_num_params = 4 * 4 + 4   # W1 and b1 parameters
L2_num_params = 4 * 1 + 1   # W2 and b2 parameters
print("L1 params = ", L1_num_params, ", L2 params = ", L2_num_params  )
```

### Performance Analysis

```python
predictions = model.predict(x_test)
accuracy = tf.keras.metrics.CategoricalAccuracy()(y_test, predictions).numpy()
print(f'Accuracy: {accuracy * 100:.2f}%')

# Reference: https://www.tensorflow.org/api_docs/python/tf/keras/metrics/CategoricalAccuracy
```

```python
binaryPred = (predictions>0.5).astype(int)

for i in range(len(y_test)):
    print("Actual: %d - Pred: %d"%(y_test[i], binaryPred[i]))
```

```python
print("Generate a prediction for a single sample")
sample1 = np.array([5.1, 2.5, 3.0, 1.1]) # Class 1: Iris-versicolor
sample2 = np.array([4.4, 3.0, 1.3, 0.2]) # Class 0: Iris-setosa

sample = sample2.reshape((1,len(sample1)))
predSample = model.predict(sample)

print(f"Prediction: {predSample}")
print(f"Prediction: {(predSample>0.5).astype(int)}")
```

```python
print("Generate predictions")
predSamples = model.predict_step(x_test[:5])
predSamples = (predSamples>0.5)
print(f"\nPrediction:\n {predSamples}")
```
