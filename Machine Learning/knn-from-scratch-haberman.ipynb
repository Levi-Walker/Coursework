# K-Nearest Neighbors from Scratch

A manual Euclidean-distance and KNN implementation using the Haberman survival dataset, followed by a scikit-learn comparison.

```python
import math
```

```python
v1 = [1.2, 2, 3.8, 4.0]
v2 = [0.5, 4.5, 5.3, 1.2]
```

```python
def EuclDist(v1,v2):
    sum = 0
    for i in range(len(v1)):
        sum = sum + (v1[i]-v2[i])**2 
    return math.sqrt(sum)
```

```python
print(EuclDist(v1,v2))
```

```python
dataset = []

with open('Desktop/haberman+s+survival/haberman.data' ,'r') as data:
    for line in data.readlines():
        l = line.replace("\n", '').split(",")
        l = [int(i) for i in l]
        dataset.append(l)
print(len(dataset))
```

```python
import random as rd
rd.shuffle(dataset)
```

```python
def info_dataset(data):
    label1, label2 = 0, 0
    for sample in data:
        if sample[-1] == 1:
            label1 += 1
        else:
            label2 +=1
    return len(data), label1, label2
```

```python
p = 0.6

total, label1, label2 = info_dataset(dataset)
```

```python
train = []
test = []

for i in range(len(dataset)):
    if i < p*total:
        train.append(dataset[i])
    else:
        test.append(dataset[i])
```

```python
info_dataset(train)
```

```python
info_dataset(test)
```

```python
def knn(train, newSample, k):
    dist = {}
    trainSize = len(train)

    for i in range(trainSize):
        d = EuclDist(train[i][:-1], newSample)
        dist[i] = d

    kNearest = sorted(dist, key=dist.get)[:k]
    nlabel1, nlabel2 = 0,0

    for ind in kNearest:
        if train[ind][-1] == 1:
            nlabel1 += 1
        else:
            nlabel2 += 1

    if nlabel1 > nlabel2:
        return 1
    else:
        return 2
    
```

```python
newSample = dataset[186][:-1]
k=7
print(dataset[186])

knn(train, newSample, k)
```

```python
k=11
hits = 0

for sample in test:
    sampleClass = knn(train,sample,k)
    if(sample[-1]) == sampleClass:
        hits +=1

print("Total training: %d"%len(train))
print("total test: %d"%len(test))

print("Correct: %.2f%%"%(100*hits/len(test)))
print("Accuracey: %.2f"%(hits/len(test)))
```

```python
# scikit-learn implementation
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=11)
```

```python
import numpy as np
```

```python
newdataset = np.array(dataset)
X = newdataset[:,:-1]
y = newdataset[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, y, train_size=0.6
)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
print("Test set predictions:\n {}".format(y_pred))
```

```python
print("Test set score: %.2f"%(np.mean(y_pred == y_test)))
```
